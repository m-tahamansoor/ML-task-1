Mean Imputation

```python
df['imputed_mean_age'] = df['Age'].fillna(df['Age'].mean())
df
```

Medium Imputation

```python
df['imputed_medium_age'] = df['Age'].fillna(df['Age'].median())
df
```

Mode Imputation

```python
df['imputed_mode-YearOfExperience'] = df['Year of Experience'].fillna(df['Year of Experience'].mode()[0])
df
```

Removing Rows

```python
removing_rows = df.dropna()
removing_rows
```

Forward Fill

```python
df['Project_Forward_Fill'] = df['Projects'].fillna(method='ffill')
df
```

Backward Fill

```python
df['Salary_Backward_Fill'] = df['Salary'].fillna(method='bfill')
df
```
